```python
import numpy as np  #multidimensional array
import pandas as pd  #cleaning data
from sklearn.model_selection import train_test_split #splitting the data set into training and test
from sklearn.linear_model import LogisticRegression #our ml model
from sklearn.metrics import accuracy_score    #accuarcy of prediction
```

```python
# loading the dataset to a Pandas DataFrame
credit_card_data = pd.read_csv('creditcard[1].csv')
```

```python
# first 5 rows of the dataset
credit_card_data.head()
```

```python
# dataset informations
credit_card_data.info()
```

```python
# checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# distribution of legit transactions & fraudulent transactions
credit_card_data['Class'].value_counts()
```

```python
# separating the data for analysis(1 for fraudulent, 0 for legit)
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data, to understand the data distribution, like its average , mean etc
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=492)
#randomly selects 492 samples of the data set
#this is basically done to balaance the data, like in case the number of non spam is much greater than spam,
#then it balances the two classes, to improve the performance of the ml model
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
#this means that a new dataset will be created by conactenating the legit data and the fraud data
#concatenation is done vertically
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class', axis=1) #creating the target matrix and output matrix
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#20%  data for testing, random state means that evrytime the code runs, splitting is still same
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
